```python
class Dial:
    def __init__(self, handler=None):
        start_pos = 50
        self.state = 0
        self.handler = handler
        self.position = start_pos

    def parseLine(self, line):
        # line wil be in format "L36" or "R23"
        direction = line[0]
        amount = int(line[1:])
        if direction == "L":
            self.moveL(amount)
        else:
            self.moveR(amount)
    
    def _move(self, amount=0):
        # handle movement in arbitrary directions.
        self.position = (self.position + amount) % 100
        if self.handler:
            self.state = self.handler(self.position, self.state)


    def moveL(self, amount=0):
        for i in range(amount):
            self._move(-1)
        #self._move(-amount)

    def moveR(self, amount=0):
        for i in range(amount):
            self._move(1)
        #self._move(amount)
```

```python
amount_0 = 0
def handler(position, state=0):
    if position == 0:
        state = state + 1
    return state
```

```python
# read the file example.txt
dial = Dial(handler)
with open("input.txt", "r") as f:
    for line in f:
        dial.parseLine(line)
print(dial.state)
```
